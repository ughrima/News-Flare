```python
import pandas as pd
from transformers import pipeline

# Load CSV
df = pd.read_csv('/content/processed_articles_with_dynamic_topics.csv')

# Text generation pipeline
generator = pipeline('text2text-generation', model='google/flan-t5-base', device=0)

# Create prompts
headline_prompts = [
    f"Write a funny, satirical headline for this news title:\n\n{title}\nFunny Headline:"
    for title in df['Title']
]
summary_prompts = [
    f"Write a 2-line sarcastic, funny summary for this article:\n\n{summary}\nFunny Summary:"
    for summary in df['Summary']
]

# Generate satirical headlines
satirical_headlines = []
for prompt in headline_prompts:
    output = generator(
        prompt,
        max_new_tokens=30,  # only new tokens will be counted
        do_sample=True,
        temperature=0.95,
        top_p=0.95,
        truncation=True  # ensure input is truncated if too long
    )[0]['generated_text']
    satirical_headlines.append(output.split('Funny Headline:')[-1].strip())

# Generate funny summaries
funny_summaries = []
for prompt in summary_prompts:
    output = generator(
        prompt,
        max_new_tokens=60,
        do_sample=True,
        temperature=0.95,
        top_p=0.95,
        truncation=True
    )[0]['generated_text']
    funny_summaries.append(output.split('Funny Summary:')[-1].strip())

# Update DataFrame
df['Satirical_Headline'] = satirical_headlines
df['Funny_Short_Summary'] = funny_summaries

# Save
df.to_csv('news_with_satirical_content.csv', index=False)

print("✅ Done! Saved to news_with_satirical_content.csv")
```
